$f(z) = z^2 + c$ $where$  $c=1-golden ration$

```python
# Declaring variables
x = linspace(-2,2,2000)
y = linspace(-2,2,2000)
xx,yy = meshgrid(x,y)
z = xx + yy * 1j
gld = (1 + sqrt(5)) / 2
c = 1 - gld
n = 200
l = zeros(z.shape)

# Performing iterations
for i in range(n): 
    z = (z ** 2) + c
    ab = abs(z) >= 2
    l[ab] = i    

# Visualizing
imshow(l);
```
